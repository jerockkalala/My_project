4. Assuming you installed Tesseract, use pytesseract to read the Bowers text image. Then, format the resulting text by removing the extraneous characters. Finally, use spaCy to print out the tokens (the text, part of speech, and dependency). (HINT: you may need to print the tokens to see what characters you need to remove.)

```python
from PIL import Image
from pytesseract import pytesseract

img = Image.open("E:\\Bellevue\\Winter_2022\\DSC 360 Data Mining Text Analytics an\\Week_3\\bowers.jpg")

#print the image, since we don't need to display the image, I put the command as a comment
print(type(img))
img.show()
```

```python
#Define path to tesseract.exe
path_to_tesseract = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

#Point tesseract_cmd to tesseract.exe
pytesseract.tesseract_cmd = path_to_tesseract
```

```python
#Extarct text from image
text = pytesseract.image_to_string(img)
print(text)
```

```python
import spacy
#print tokens
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
for token in doc:
    print(token.text)
```

#Now let remove special character from our text

```python
import re

# Removing special characters, Only keep:
# Letters a-z, A-Z
# Dots (.)
# Numbers (\d)
# Non white space (\s)
# ^ means other than the provided patterns
# need to be substituted by an empty string
text3 = re.sub('[^a-zA-Z.\d\s]', '', text)

# Print the final string. This should print:
print(text3)
```

```python
#print Token of the new text without special characters
doc = nlp(text3)
for token in doc:
    print(token.text)
```

```python
#Part Of Speech and dependencies
for token in doc:
    print(f'{token.text:{10}} {token.pos_:{10}} {token.tag_:{10}} {spacy.explain(token.tag_)}')
```
